## Observations and Insights 

The size of the mouse has a strong effect on whether or not the treatment is effective. Mouse weight and average tumor size have a strong correlation coefficient of 0.84.

The results of the four treatments produce fairly reliable results. Only one result out of 247 mice was an outlier outside of the interquartile range for their treatment.

Visually looking at the boxplots for the treatments, the drugs Capomulin and Ramicane seem to be more effective than Infubinol or Ceftamin. The upper bounds of the interquartile ranges for the two former are comparable to the lower bounds of the latter two.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata,study_results,how='outer', on='Mouse ID')
# Display the data table for preview
combined_df.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = combined_df.set_index("Mouse ID")
cleaned_df = cleaned_df.drop(index="g989")
cleaned_df = cleaned_df.reset_index()
cleaned_df.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
tumor_group = cleaned_df.groupby("Drug Regimen")["Tumor Volume (mm3)"]

means = tumor_group.mean()
medians = tumor_group.median()
variances = tumor_group.var()
deviations = tumor_group.std()
sems = tumor_group.sem()

summary_df = pd.DataFrame({"Mean":means,"Median":medians,"Variance":variances,"Standard Deviation":deviations,"SEM":sems})
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_tp_df = pd.DataFrame(cleaned_df.groupby("Mouse ID")["Timepoint"].max())

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_size_df = pd.merge(final_tp_df,cleaned_df, on=["Mouse ID", "Timepoint"], how="left")
final_size_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    temp_df = final_size_df.loc[final_size_df["Drug Regimen"]==treatment]
    
    # add subset 
    tumor_data.append(temp_df["Tumor Volume (mm3)"])
    
    # Determine outliers using upper and lower bounds
    quartiles = temp_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    print(f"The lower quartile of {treatment} is: {lowerq}")
    print(f"The upper quartile of {treatment} is: {upperq}")
    print(f"The interquartile range of {treatment} is: {iqr}")
    print(f"The the median of {treatment} is: {quartiles[0.5]} ")

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    
    outliers_count = temp_df.loc[(temp_df["Tumor Volume (mm3)"] > upper_bound) 
                                 | (temp_df["Tumor Volume (mm3)"] < lower_bound),"Mouse ID"].count()
    print(f"Outlier count for {treatment} is: {outliers_count}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
props = dict(marker='o',markerfacecolor='red',markeredgecolor='black')
plt.boxplot(tumor_data, flierprops=props)
plt.ylabel("Final Tumor Volume (mm3)")
plt.xlabel("Drug Regimen")
plt.xticks(ticks=np.arange(1,5),labels=treatments)
plt.title("Final Tumor Volume by Drug Regimen")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
mouse = cleaned_df.loc[cleaned_df["Mouse ID"] == "i557"]

plt.plot(mouse["Timepoint"], mouse["Tumor Volume (mm3)"], marker='|', markeredgecolor='black', ms=8)
plt.title("Mouse i557 Tumor Size Over Time")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.xlim(0,45)
plt.ylim(42,48)
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capo = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Capomulin"]
average_df = pd.DataFrame(capo.groupby("Mouse ID")["Tumor Volume (mm3)"
                                                  ].mean()).rename(columns={"Tumor Volume (mm3)":"AVG Tumor Volume"})
average_merge_df = pd.merge(average_df,capo.drop_duplicates("Mouse ID"), on="Mouse ID",how="outer")

plt.scatter(average_merge_df["Weight (g)"], average_merge_df["AVG Tumor Volume"])
plt.title("Average Tumor Volume vs Mouse Weight")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
plt.scatter(average_merge_df["Weight (g)"], average_merge_df["AVG Tumor Volume"])
plt.title("Average Tumor Volume vs Mouse Weight")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(average_merge_df["Weight (g)"], average_merge_df["AVG Tumor Volume"])
regress_values = average_merge_df["Weight (g)"] * slope + intercept
plt.plot(average_merge_df["Weight (g)"],regress_values,"r-")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(20,36),fontsize=15,color="red")

plt.show()

correlation = st.pearsonr(average_merge_df["Weight (g)"], average_merge_df["AVG Tumor Volume"])
print(f"The correlation coefficient is {round(correlation[0],2)}.")
```
